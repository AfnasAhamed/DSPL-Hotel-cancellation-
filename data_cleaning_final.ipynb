## Whitespace Check

```python
def check_whitespace(df, name):
    str_cols = df.select_dtypes(include='object').columns
    for col in str_cols:
        count = df[col].str.contains(r'^\s|\s$', regex=True).sum()
        if count > 0:
            print(f"{name} — {col}: {count} rows with whitespace")
```

```python
check_whitespace(train_data, 'train')
check_whitespace(validation_data, 'validation')
check_whitespace(test_data, 'test')
```

```python
str_cols = train_data.select_dtypes(include='object').columns
train_data[str_cols] = train_data[str_cols].apply(lambda col: col.str.strip())

str_cols = validation_data.select_dtypes(include='object').columns
validation_data[str_cols] = validation_data[str_cols].apply(lambda col: col.str.strip())

str_cols = test_data.select_dtypes(include='object').columns
test_data[str_cols] = test_data[str_cols].apply(lambda col: col.str.strip())
```

## Date Type Conversion

```python
cols = ['Expected_checkin', 'Expected_checkout', 'Booking_date']

for col in cols:
    train_data[col] = pd.to_datetime(train_data[col], errors='coerce')
    validation_data[col] = pd.to_datetime(validation_data[col], errors='coerce')
    test_data[col] = pd.to_datetime(test_data[col], errors='coerce')
```
