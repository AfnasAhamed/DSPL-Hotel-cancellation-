# **Seasonal Cancellation Risk for each hote**

Create Season Columns

```python
print(df.columns)
```

```python
season_map = {
    12: "Winter", 1: "Winter", 2: "Winter",
    3: "Spring", 4: "Spring", 5: "Spring",
    6: "Summer", 7: "Summer", 8: "Summer",
    9: "Autumn", 10: "Autumn", 11: "Autumn"
}

df["Season"] = df["Checkin_month"].map(season_map)
```

Reservation_Status	Cancelled_Flag


Canceled	1


Check-In	0


No-Show	0





```python
df["Cancelled_Flag"] = (df["Reservation_Status"] == "Canceled").astype(int)
```

```python
season_cancel = df.groupby(["Season", "Hotel_Type"])["Cancelled_Flag"].mean().reset_index()

season_cancel["Cancelled_Flag"] = season_cancel["Cancelled_Flag"] * 100

season_cancel
```

```python
plt.figure(figsize=(10,6))

sns.barplot(
    data=season_cancel,
    x="Season",
    y="Cancelled_Flag",
    hue="Hotel_Type"
)

plt.title("Seasonal Cancellation Risk by Hotel")
plt.xlabel("Season")
plt.ylabel("Cancellation Rate (%)")

plt.show()
```

```python
season_status = pd.crosstab(
    [df["Season"], df["Hotel_Type"]],
    df["Reservation_Status"],
    normalize="index"
) * 100

season_status = season_status.reset_index()

season_risk = season_status.melt(
    id_vars=["Season", "Hotel_Type"],
    value_vars=["Canceled", "No-Show"],
    var_name="Status",
    value_name="Percentage"
)

plt.figure(figsize=(12,6))
sns.barplot(
    data=season_risk,
    x="Season",
    y="Percentage",
    hue="Status"
)

plt.title("Seasonal Cancellation and No-Show Risk")
plt.xlabel("Season")
plt.ylabel("Percentage")
plt.show()
```

## **Proposed Deadline**

**The analysis of cancellation rates by lead time shows that cancellation risk increases noticeably after approximately 20 days before the check-in date. Bookings made far in advance are more likely to be cancelled compared to bookings made closer to the stay date. Based on this pattern, a cancellation policy with a 20-day deadline is proposed. Customers cancelling more than 20 days before arrival would incur a lower fee, while cancellations closer to the check-in date would incur a higher fee to compensate for the increased risk of lost revenue**.

# Proof for propsal

```python
plt.figure(figsize=(10,6))

plt.plot(cancel_rate.index.astype(str), cancel_rate.values*100, marker="o")

plt.axvline(x=3, color="red", linestyle="--", label="Proposed 20-day Deadline")

plt.title("Cancellation Rate by Lead Time Group")
plt.xlabel("Lead Time Group")
plt.ylabel("Cancellation Rate (%)")

plt.legend()

plt.show()
```

```python
df["Deadline_Group"] = df["Lead_Time_Days"].apply(
    lambda x: "≤20 Days" if x <= 20 else ">20 Days"
)

deadline_cancel = df.groupby("Deadline_Group")["Reservation_Status"].apply(
    lambda x: (x == "Canceled").mean()
) * 100

deadline_cancel = deadline_cancel.round(2)

deadline_cancel.plot(kind="bar", figsize=(8,5))

plt.title("Cancellation Rate Before and After 20-Day Deadline")
plt.xlabel("Lead Time Category")
plt.ylabel("Cancellation Rate (%)")

plt.xticks(rotation=0)

plt.show()
```
